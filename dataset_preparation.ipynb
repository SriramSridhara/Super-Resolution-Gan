```python
import cv2
import os
from PIL import Image

from collections import Counter
import matplotlib.pyplot as plt
```

```python
hr_path = 'C:/Users/HP/OneDrive/Desktop/Ram/SRGAN/Data/HR'
image_files = [f for f in os.listdir(hr_path) if f.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff'))]
```

```python
for i, filename in enumerate(image_files, start=1):
    # Get the file extension
    file_extension = os.path.splitext(filename)[1]
    
    # Create the new name (e.g., "1.jpg", "2.png")
    new_name = f"{i}{file_extension}"
    
    # Get full file paths
    old_path = os.path.join(hr_path, filename)
    new_path = os.path.join(hr_path, new_name)
    
    # Rename the file
    os.rename(old_path, new_path)
    print(f"Renamed: {filename} -> {new_name}")

print("Renaming completed!")
```

```python
dimensions = []

for filename in os.listdir(hr_path):
    #full path of an image
    image_path = os.path.join(hr_path, filename)
    
    with Image.open(image_path) as img:
        # Check the dimensions
        width, height = img.size
        
        # Rotate if portrait
        if height > width:
            img = img.rotate(90, expand=True)
            width, height = img.size  # Update dimensions after rotation
            img.save(image_path)
            print(f"Rotated: {filename}")
        else:
            print(f"Already landscape: {filename}")
        
        # Store the dimensions
        dimensions.append((width, height))
```

```python
#Step 2: Calculate the most common dimensions
widths, heights = zip(*dimensions)  # Split into separate lists for width and height
most_common_width = Counter(widths).most_common(1)[0][0]  # Most common width
most_common_height = Counter(heights).most_common(1)[0][0]  # Most common height

print(f"Most Common Width: {most_common_width}")
print(f"Most Common Height: {most_common_height}")

# Step 3: Plot histograms
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(widths, bins=20, color='blue', alpha=0.7)
plt.title('Width Histogram')
plt.xlabel('Width')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(heights, bins=20, color='green', alpha=0.7)
plt.title('Height Histogram')
plt.xlabel('Height')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Step 4: Resize all images to the most common dimensions
image_files = [f for f in os.listdir(hr_path) if f.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff'))]
for filename in image_files:
    image_path = os.path.join(hr_path, filename)
    
    with Image.open(image_path) as img:
        resized_image = img.resize((most_common_width, most_common_height))
        resized_image.save(image_path)  # Overwrite the original file
        print(f"Resized: {filename} to {most_common_width}x{most_common_height}")

print("All images resized to the most common dimensions!")
```

Now to generate lower resolution images

```python
lr_path = 'C:/Users/HP/OneDrive/Desktop/Ram/SRGAN/Data/LR'


# Downsample and save images
for filename in os.listdir(hr_path):
    input_path = os.path.join(hr_path, filename)
    
    # Open the image
    with Image.open(input_path) as img:
        # Calculate new dimensions
        new_width = int(img.width * 0.5)
        new_height = int(img.height * 0.5)
        
        # Resize the image
        downsampled_img = img.resize((new_width, new_height), Image.Resampling.LANCZOS)

        
        # Prepare the new filename
        name, ext = os.path.splitext(filename)
        new_filename = f"{name}_lr{ext}"  # Add '_lr' to the original name
        output_path = os.path.join(lr_path, new_filename)
        
        # Save the downsampled image
        downsampled_img.save(output_path)
        print(f"Saved: {new_filename}")

print("Downsampling completed! All images saved in:", lr_path)
```
